```python
autos.rename({'yearOfRegistration':'registration_year',
             'monthOfRegistration':'registration_month',
             'notRepairedDamage':'unrepaired_damage',
             'dateCreated':'ad_created'},
            axis=1, inplace=True)
```

First, I just changed a couple of the column names to make them more clear, readable, and easy to use. Now to determine what other cleaning is needed.

```python
autos.describe(include='all')
```

The price and odometer columns are text, but should be numbers. I'll clean those up and make them numeric next as well as rename the odometer column to make it clear it is in kilometers.

Additionally, some columns have no variability whatsoever. For instance, there are two options for seller, and 49,999 out of 50,000 are the same (private). The same is true for offer type.

```python
# Get rid of the text
autos['odometer'] = (autos['odometer']
                     .str.replace('km', '')
                     .str.replace(",", '')
                     .astype(float)
                    )
autos['price'] = (autos['price']
                  .str.replace('$', '')
                  .str.replace(',', '')
                  .astype(float)
                 )

# Rename the column
autos.rename({'odometer':'odometer_km'}, axis=1, inplace=True)
```

How old are the cars?

Range between about 1999 and 2008, average year being 2004. Though, this could be pulled a bit by the clear non-true values (minimum of 1000 and max of 9999). So I'll quickly just take a look at the frequencies, decide my reasonable upper and lower bound, and remove any cars outside that range.

```python
autos['registration_year'].value_counts().sort_index()
```

Now, the first automobile in existence was built in 1885, so based on these, the lowest plausible year is 1910 so I'll keep that And given the data were crawled in early 2016, values that are 2017 and above are also not plausible. So I'll just keep cars from 1910 to 2016.

```python
autos = autos[autos['registration_year'].between(1910,2016)]

# Logic check to make sure we included the boundary years
(autos['registration_year']
.value_counts(normalize=True)
.sort_index())
```

There we go. Most cars fall between 1994 and 2016.

## Brand prices

Going to do a bit of aggregation on brand, looking at average prices.

```python
autos['brand'].unique().tolist()
```

```python
freqs = (autos['brand']
 .value_counts(normalize=True))
freqs
```

Not a ton of brands. Let's just take the most popular.

Looking at this, I'll take from VW to Hyundai. This will get us a good chunk of all the most popular cars. Also, there are some Lada's on there. Which is hilarious.

```python
mean_prices = {}

brands = autos['brand'].unique()
for b in brands:
    selected_rows = autos[autos['brand'] == b]
    mean_price = selected_rows['price'].mean()
    mean_prices[b] = mean_price
    
```

```python
mean_prices
```

Doing the same as above to get another dictionary of average mileage, so we can compare price to mileage - is there a link?

```python
mean_kms = {}

for b in brands:
    selected_rows = autos[autos['brand'] == b]
    mean_km = selected_rows['odometer_km'].mean()
    mean_kms[b] = mean_km
    
# Turn both dictionaries into series
price_series = pd.Series(mean_prices)
kms_series = pd.Series(mean_kms)

# Make a dataset
price_kms_data = pd.DataFrame(price_series, columns=['mean_price'])
price_kms_data['mean_kms'] = kms_series

# Logic Check
price_kms_data
```

```python
# quick Scatter plot to see if there is a relationship
import matplotlib.pyplot as plt
% matplotlib inline
plt.style.use('fivethirtyeight')

plt.scatter(x=price_kms_data['mean_kms'],
           y=price_kms_data['mean_price'])
plt.ylim(0,40000)
plt.title('Plot of Average Price and Odometer \nby Brand')
plt.xlabel('Average Odometer Reading (kms)')
plt.ylabel('Average Price ($)')
plt.show()
```

Seems to be a small downward trend with more kilometers meaning lower average price, but it's hard to tell!
